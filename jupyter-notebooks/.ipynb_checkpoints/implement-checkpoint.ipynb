```python
def get_aqi_range(aqi):
    if aqi <= 50:
        return 'Good', 'Zone IV'
    elif 50 < aqi <= 100:
        return 'Moderate', 'Zone IV'
    elif 100 < aqi <= 150:
        return 'Unhealthy for sensitive groups', 'Zone III'
    elif 150 < aqi <= 200:
        return 'Unhealthy', 'Zone III'
    elif 200 < aqi <= 300:
        return 'Very Unhealthy', 'Zone II'
    else:
        return 'Hazardous', 'Zone I'
```

### Score function
T(i) = tolerance, U(i) = Utility

Score(i) = w1 * T(i) + w2 * U(i)

```python
def get_score(zone, tree_data=tree_data, w1=10, w2=5):
    score = [0]*len(tree_data)
    area = [0]*len(tree_data)
    cost = [0]*len(tree_data)
    for i in range(len(tree_data)):
        # f = w1*pi + w2*ui
        score[i] = w1*tree_data[i][zone] + w2*tree_data[i]['Utility']
        area[i] = tree_data[i]['Area']
        cost[i] = tree_data[i]['Cost']
    return score, area, cost
```

```python
score, area, cost = get_score(Zone)
print(score)
print(area)
print(cost)
```

# objective function
F = w1 * Σscore(i) / N  +  w2 * (Σarea(i) - A)  +  w3 * (Σcost(i) - C)


w1, w2, w3 are weights

N = population

A = Area available

C = Cost Budget available

```python
# objective funtion
def calculate_fitness(total_score, total_cost, total_area,
                      C, A, population, w1=1000, w2=1, w3=0.01):
    f = w1*total_score/population + w2*(total_area - A) + w3*(total_cost - C)
    return f
```

```python
# first create a sample set
def get_sampling_set(area, cost, A, C):
    S = []
    minc, maxc = float('inf'), -float('inf')
    for i in range(len(area)):
        count = min(C//cost[i], A//area[i])
        minc = min(count, minc)
        maxc = max(count, maxc)
        S.extend([i]*count)
    return S, minc, maxc
```

```python
sampling_set, minc, maxc = get_sampling_set(area, cost, A, C)
print(sampling_set, minc, maxc, sep='\n')
```

```python
X = 20    # no of chromosomes
```

```python
def initialize_chromosomes(X, sample_size, minc, maxc):
    chromosomes = [[0]*sample_size for i in range(X)]
    for c in chromosomes:
        choose = random.randint(minc, maxc)
        idx = random.sample(range(len(c)), choose)
        for i in idx:
            c[i] = 1
    return chromosomes
```

```python
chromosomes = initialize_chromosomes(X, len(sampling_set), minc, maxc)
for c in chromosomes:
    print(*c, sep='')
```

```python
def get_fitness(chromosome, score=score, sampling_set=sampling_set,
                cost=cost, area=area, C=C, A=A, population=population):
    N = len(chromosome)
    total_cost = sum(cost[sampling_set[i]] for i in range(N) if chromosome[i])
    total_area = sum(area[sampling_set[i]] for i in range(N) if chromosome[i])
    if total_cost > C or total_area > A:
        return -float('inf')
    total_score = sum(score[sampling_set[i]] for i in range(N) if chromosome[i])
    total_fitness = calculate_fitness(total_score, total_cost, total_area, C, A, population)
    return total_fitness
```

```python
tot_fit = []
for c in chromosomes:
    tot_fit.append(get_fitness(c))
    print(*c, "\t", tot_fit[-1], sep='')
```

```python
def perform_crossovers(chromosomes, tot_fit):
    M = len(chromosomes[0])    # length of a chromosome
    X = len(chromosomes)       # no. of chromosomes
    sorted_chrom = [i[0] for i in sorted(zip(chromosomes, tot_fit),
                                         key=lambda i:i[1], reverse=True)][:X//2]
    for i in range(X//2):
        pivot = random.randint(1, M-2)
        # array modified in place
        chromosomes[i] = sorted_chrom[i][:pivot] + sorted_chrom[X//2-i-1][pivot:]
        chromosomes[X-i-1] = sorted_chrom[X//2-i-1][:pivot] + sorted_chrom[i][pivot:]
        chromosomes[X//2-i-1] = sorted_chrom[i]
        chromosomes[X//2+i] = sorted_chrom[X//2-i-1]
```

```python
perform_crossovers(chromosomes, tot_fit)

tot_fit = []
for c in chromosomes:
    tot_fit.append(get_fitness(c, score, sampling_set, cost, area, C, A))
    print(*c, "\t", tot_fit[-1], sep='')
```

```python
def assign_fitness(chromosome, tot_fit, i):
    tot_fit[i] = get_fitness(chromosome)
```

```python
from threading import Thread
from math import isclose
```

```python
def run_genetic_algorithm(chromosomes, X, max_iter = 1000, max_rep = 20):
    rep_count = 0
    threads = [None]*X
    best_ch = None
    best_fit = -float('inf')
    tot_fit = [0]*X
    prev_fit = -float('inf')

    for t in range(max_iter):
        for i in range(X):
            threads[i] = Thread(target=assign_fitness, args=(chromosomes[i], tot_fit, i))
            threads[i].start()
        for th in threads:
            th.join()

        curr_best_fit, curr_best_ch = min(zip(tot_fit, chromosomes), key=lambda i:i[0])
        if curr_best_fit > best_fit:
            best_fit = curr_best_fit
            best_ch = curr_best_ch

        if t%100 == 0:
            print(curr_best_fit)

        if isclose(curr_best_fit, prev_fit, rel_tol=0.1):
            rep_count += 1
        prev_fit = curr_best_fit

        if rep_count >= max_rep:
            rep_count = 0 
            chromosomes = initialize_chromosomes(X, len(sampling_set), minc, maxc)
            continue
        perform_crossovers(chromosomes, tot_fit)
    
    return best_ch, best_fit
```

```python
best_ch, best_fit = run_genetic_algorithm(chromosomes, X, max_iter=3000)
print(best_ch, best_fit)
```

```python
from collections import defaultdict
```

```python
def collect_idx_counts(chromosome, sampling_set, tree_data):
    trees = defaultdict(int)
    for i in range(len(sampling_set)):
        if chromosome[i]:
            trees[tree_data[sampling_set[i]]['Common name']] += 1
    return dict(trees)
```

```python
trees = collect_idx_counts(best_ch, sampling_set, tree_data)
trees
```

```python
tot_score = sum(score[tree_idx[t]]*trees[t] for t in trees)
tot_area = sum(area[tree_idx[t]]*trees[t] for t in trees)
tot_cost = sum(cost[tree_idx[t]]*trees[t] for t in trees)
tot_score, tot_area, tot_cost
```
